```python
# Importing the libraries  
import numpy as nm  
import matplotlib.pyplot as py  
import pandas as pd  
```

```python
# Importing the dataset  
dataset = pd.read_csv('Mall_Customers.csv')  
```

```python
x = dataset.iloc[:, [3, 4]].values 
```

```python
#Finding the optimal number of clusters using the dendrogram  
import scipy.cluster.hierarchy as shc  
dendro = shc.dendrogram(shc.linkage(x, method="ward"))  
py.title("Dendrogram Plot")  
py.ylabel("Euclidean Distances")  
py.xlabel("Customers")  
py.show() 
```

```python
#training the hierarchical model on dataset  
from sklearn.cluster import AgglomerativeClustering  
hc= AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')  
y_predict= hc.fit_predict(x) 
labels= hc.labels_
```

```python
from sklearn import metrics
print(metrics.silhouette_score(x, labels, metric='euclidean'))
```

```python
print(metrics.calinski_harabasz_score(x, labels))
```

```python
#visulaizing the clusters  
py.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1], s = 100, c = 'black', label = 'Cluster 1')  
py.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1], s = 100, c = 'orange', label = 'Cluster 2')
py.scatter(x[y_predict== 2, 0], x[y_predict == 2, 1], s = 100, c = 'magenta', label = 'Cluster 3')
py.scatter(x[y_predict == 3, 0], x[y_predict == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4') 
py.scatter(x[y_predict == 4, 0], x[y_predict == 4, 1], s = 100, c = 'red', label = 'Cluster 5')
py.title('Clusters of customers')  
py.xlabel('Annual Income ($)')  
py.ylabel('Spending Score (1-100)')  
py.legend()  
py.show()
```
